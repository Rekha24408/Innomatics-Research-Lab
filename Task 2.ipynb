Problem 2 :

```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
highest_employee = max(employees, key=employees.get)
print("Highest Salary:", highest_employee, "-", employees[highest_employee])
```

Problem 3:

```python
numbers = [45, 22, 89, 10, 66]
max_val = numbers[0]
min_val = numbers[0]
for num in numbers:
    if num > max_val:
        max_val = num
    if num < min_val:
        min_val = num
print("List:", numbers)
print("Max:", max_val)
print("Min:", min_val)
```

Problem 4:

```python
prices = [450, 1200, 899, 1500, 300]
count = 0
for price in prices:
    if price > 1000:
        count += 1
print("Products above 1000:", count)
```

Problem 5:

```python
Attendance = ["P", "P", "A", "P", "P"]
present_count = Attendance.count("P")
total_days = len(Attendance)
percentage = (present_count / total_days) * 100
print("Attendance Percentage:", percentage)
```

Problem 7:

```python
text = "pythonp"
char_count = {}
for char in text:
    if char in char_count:
        char_count[char] += 1
    else:
        char_count[char] = 1
print(char_count)
```
